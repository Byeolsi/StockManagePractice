```python
df = fs.fn_consolidated(otp='16479279991001081001', term='2020Q3', vol=100000, study='Y')
df.head(3)
```

```python
df.columns
```

```python
df['PER'] = df['Price'] / df['EPS']
df['PER']
```

```python
terms = fs.set_terms(trade_start='2006Q2', trade_end='2021Q2') # trade term to fiscal terms
terms
```

```python
data = {}
for t in terms:
  data[t] = fs.fn_consolidated(otp='16479279991001081001', term=t)
```

```python
for t in terms:
  data[t]['PER'] = data[t]['Price'] / data[t]['EPS']
```

```python
s = {}
signal = {}
for t in terms:
  s[t] = fs.fn_filter(data[t], by='PER', floor=1, cap=10, n=30, asc=True) # fn_filter 를 사용해 종목 선정
  signal[t] = list(s[t].index) # 선정된 종목을 signal 에 저장
  print(signal[t])
```

```python
df = fs.backtest(signal=signal, data=data, m=2, cost=.001)
```

```python
fs.draw_return(df)
```

```python
df = fs.fn_single(otp='16479279991001081001', symbol='JBLU', window='T')
```

```python
fs.draw_price_multiple_band(df, multiple='PBR', acct='BPS') # PBR 밴드 그려보기
```

```python
terms = fs.set_terms(trade_start='2006Q2', trade_end='2021Q2')
data = {}
for t in terms:
  data[t] = fs.fn_consolidated(otp='16479279991001081001', term=t)
```

```python
s = {}
signal = {}
for t in terms:
  data[t]['BPS'] = data[t]['Shareholders Equity'] / data[t]['Shares']
  data[t]['PBR'] = data[t]['Price'] / data[t]['BPS']
  s[t] = fs.fn_filter(data[t], by='PBR', floor=.1, cap=2, n=30, asc=True)
  signal[t] = list(s[t].index)

df = fs.backtest(signal=signal, data=data, m=2, cost=.001)
```

```python
fs.draw_return(df)
```
